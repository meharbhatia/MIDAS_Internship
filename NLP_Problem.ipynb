### Tranfer Learning and Universal Sentence Encoders

‘Universal Embeddings’ which are basically pre-trained embeddings obtained from training deep learning models on a huge corpus. This enables us to use these pre-trained (generic) embeddings in a wide variety of tasks including, scenarios with constraints like lack of adequate data. This is a perfect example of transfer learning, leveraging prior knowledge from pre-trained embeddings to solve a completely new task!

### Function to read the csv file 

```python
#Function to read the csv file
def read_csv(data_path):
    file_reader = csv.reader(open(data_path,"rt", errors="ignore",encoding="utf-8"), delimiter=',')
    sent_list = []

    for row in file_reader:
        id = row[0]
        sent = row[1]
        lab=row[2]
        sent_list.append(np.array([id,sent,lab]))
    return np.stack(sent_list)
```

### Preprcessing 
- In the training data, need to remove the " " at the beginning and end of the sentence. 
- Here is a function to clean the text in the training data

```python
def cleanData(sentence):
    if sentence[0] == '"' and sentence[-1] == '"':
        return sentence[1:-1]
    else:
        return sentence
```

### Time to use the Universal Sentence Encoder from TFHub

```python
url = 'https://tfhub.dev/google/universal-sentence-encoder-large/3'
embed = hub.Module(url)
```

### Need the help of a function to replace labels with one-hot encoders

```python
def oneHotEncode(y):
    return keras.utils.to_categorical(y)
```

### Separating training testing and evalution datasets into text and labels

```python
x_train = train[1]
y_train = train[2]
y_train_cat = oneHotEncode(y_train)

x_test = test['sentence']
y_test = test['label']
y_test_cat = oneHotEncode(y_test)

x_eval = evaluation[1]
```

### Lower case the text to make the Universal Sentence Encoder easier to use

```python
x_train = x_train.str.lower()
x_test = x_test.str.lower()
x_eval = x_eval.str.lower()
```

### Creating model with 4 dense layers and two Dropout layers

```python
from keras.layers import Dense, Dropout
from keras.models import Sequential
import keras.backend as K


def create_model():
    model = Sequential([
      
      Dense(512, activation='relu', input_dim=512),
      Dense(256, activation='relu'),
      Dropout(0.25),
      Dense(64, activation='relu'),
      Dropout(0.2),
      
      Dense(2, activation='softmax')
    ])
  
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
def get_embeddings(data):
    with tf.Session() as session:
        session.run([tf.global_variables_initializer(), tf.tables_initializer()])
        embeddings = session.run(embed(data))
        return embeddings
```

```python
x_train_splits = np.array_split(x_train.values, 17)
x_train_splits = np.array(x_train_splits)
```

```python
train_embeddings = []

for split in x_train_splits:
    embedding = get_embeddings(split)
    train_embeddings.append(embedding)
```

```python
train_embeddings = np.array(train_embeddings)
train_embeddings = train_embeddings.reshape(-1, train_embeddings.shape[-1])
train_embeddings.shape
```

```python
model = create_model()
```

```python
model.summary()
```

```python
model.fit(train_embeddings, y_train_cat, validation_split=0.1, epochs=5, batch_size=500)
```
